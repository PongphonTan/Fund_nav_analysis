```python
import pandas as pd
import glob
import seaborn as sns
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
plt.style.use('ggplot')
%matplotlib inline
```

# importing data in CSV into consolidated dataframe

```python
path = r"C:\Users\Family\OneDrive - Mahidol University\PT\script\Fund_data" # use your path
all_files = glob.glob(path + "\*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    print(f'''Readed = {filename}''')
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)

frame.head()
```

```python
frame.iloc[:,15].value_counts()
```

## dorp not required field

```python
#original_data = frame.drop(columns=['nameTh'])
original_data = frame[['navDate', 'nameEn', 'navPerUnit']]

original_data.head()
```

# Select only requried date and order by date

```python
original_data = original_data[ (frame['navDate'] >= '2018-01-01T00:00:00+07:00') & (frame['navDate'] <= '2023-02-28T00:00:00+07:00') ]

original_data.sort_values(by='navDate', inplace = True)

original_data
```

# check date time range in data set

```python
print(f'''frame : start = {frame['navDate'].min()}, end = {frame['navDate'].max()}''')
print(f'''original_data : start = {original_data['navDate'].min()}, end = {original_data['navDate'].max()}''')
```

# Data cleanning

## define data types

```python
original_data = original_data.astype({'navDate': 'datetime64[ns, ASIA/Bangkok]', 'nameEn': 'string'})
original_data.dtypes
```

```python
original_data['navDate'].value_counts()
```

## clean date

```python
original_data['navDate_year'] = original_data['navDate'].dt.year
original_data['navDate_month'] = original_data['navDate'].dt.month

original_data['navMonth'] = original_data['navDate'].dt.to_period('M').dt.to_timestamp()

original_data
```

# Select ATH of navPerUnit and corresponding Date per Fund

```python
Fund_max_nav_detail = original_data.sort_values('navPerUnit', ascending=False).drop_duplicates(['nameEn'])

Fund_max_nav_detail
```

```python
Fund_max_nav = pd.DataFrame(original_data.groupby(['nameEn'], as_index=False)['navPerUnit'].max()).sort_values(by=['navPerUnit'], ascending=False)

Fund_max_nav
```

# navPerUnit by Fund for each month

```python
Fund_Top_Month = pd.DataFrame(original_data.sort_values('navPerUnit', ascending=False).groupby(['navMonth', 'nameEn'], sort=False, as_index=False)['navPerUnit'].max())

Fund_Top_Month
```

```python
Fund_timeseries = pd.DataFrame(original_data.sort_values('navPerUnit', ascending=False).groupby(['navDate', 'nameEn'], sort=False, as_index=False)['navPerUnit'].max())

Fund_timeseries
```

# list of top fund by max navPerUnit

```python
#Fund_timeseries_top = Fund_timeseries.sort_values('navPerUnit', ascending = False).groupby('nameEn').head(1)
Fund_name_top = Fund_timeseries.sort_values('navPerUnit', ascending = False).groupby('nameEn').head(1)['nameEn'][:20].tolist()

Fund_name_top
```

```python
Fund_timeseries_top = Fund_timeseries[Fund_timeseries['nameEn'].isin(Fund_name_top)]

Fund_timeseries_top
```

# Change of NAV

```python
NAV_change = frame[['navDate', 'nameEn', 'change']]

NAV_change
```

```python
NAV_change = original_data[ (frame['navDate'] >= '2018-01-01T00:00:00+07:00') & (frame['navDate'] <= '2023-02-28T00:00:00+07:00') ]

NAV_change.sort_values(by='navDate', inplace = True)

NAV_change
```

```python
print(f'''frame : start = {frame['navDate'].min()}, end = {frame['navDate'].max()}''')
print(f'''original_data : start = {NAV_change['navDate'].min()}, end = {NAV_change['navDate'].max()}''')
```

```python
NAV_change = original_data.astype({'navDate': 'datetime64[ns, ASIA/Bangkok]', 'nameEn': 'string'})
NAV_change.dtypes
```

```python
NAV_change['navMonth'] = NAV_change['navDate'].dt.to_period('M').dt.to_timestamp()
NAV_change['navWeek'] = NAV_change['navDate'].dt.to_period('w').dt.to_timestamp()

NAV_change
```

## Aggregate to monthly/weekly change using max

```python
NAV_change_agg_max = pd.DataFrame(NAV_change.sort_values('change', ascending=False).groupby(['navWeek', 'nameEn'], sort=False, as_index=False)['change'].max())

NAV_change_agg_max
```

```python
NAV_change_agg_max_filter = NAV_change_agg_max[NAV_change_agg_max['nameEn'].isin(
    ['The ABF Thailand Bond Index Fund',
    'SIAM SELECT FUND',
    'SUBSOMBOON FUND',
    'KA Equity Fund',
    'abrdn Growth Fund',
    'BCAP Global Private Equity 22A Fund Not for Retail Investors',
    'ONE STOXX ASEAN  SELECT DIVIDEND INDEX FUND',
    'TMB SET50',
    'TISCO Aggressive Growth Fund',
    'Bualuang Equity RMF',
    'Equity Retirement Mutual Fund',
    'abrdn Smart Capital Retirement Mutual Fund',
    'K Flexible Equity RMF',
    'UOB EQUITIES RMF',
    'VALUED STOCK RETIREMENT MUTUAL FUND',
    'TISCO Equity Growth Retirement  Fund',
    'UOB SMART MILLENNIUM GROWTH FUND',
    'K Flexible Equity Fund',
    'MFC Retirement Value Fund',
    'TISCO Strategic Fund']
)]
```

```python
plt.figure(figsize=(15,6))
sns.set_style('whitegrid')
sns.set(rc={"xtick.bottom" : True, "ytick.left" : True})
ax2 = sns.lineplot(x = "navWeek", y = "change", data=NAV_change_agg_max_filter, hue="nameEn", palette = "Set1")
ax2.set_title('Monthly Change of NAV per unit (Baht)')
ax2.yaxis.get_ticklocs(minor=True)
ax2.minorticks_on()
#ax.xaxis.set_tick_params(which='minor', bottom=False)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5)) # Add legend outside of the plot 
```
